```python
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
```

```python
data = pd.read_csv("PSB/organizations.csv")
data = data.drop(columns=["invalid_info", "Unnamed: 0", "Unnamed: 0.1"],  axis=1)

data['company_type'] = data['name'].str.extract(r'^(ООО|ЗАО|ПАО|АО|ИП|ОАО)')
data['name'] = data['name'].str.replace(r'^(ООО|ЗАО|ПАО|АО|ИП|ОАО)\s*', '', regex=True)

status_dummies = pd.get_dummies(data['status'], prefix='status').astype(int)
data = pd.concat([data, status_dummies], axis=1)
data = data.drop('status', axis=1)

data.to_csv("PSB/organizations_processed.csv", index=False)
data.head(30)
```
