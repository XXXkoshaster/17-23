```python
import pandas as pd
import json

pd.set_option('display.max_info_columns', 200)
```

```python
data = pd.read_csv('PSB/data/finances.csv')
inns = data['inn']
data['data'] = data['data'].apply(json.loads)
```

```python
df_expanded = pd.json_normalize(data['data'])
df_expanded.columns = [col.replace('.', '_') for col in df_expanded.columns]
```

```python
df_expanded = pd.concat([inns, df_expanded], axis=1)
df_expanded.to_csv("PSB/data/finances_processed.csv")
```
